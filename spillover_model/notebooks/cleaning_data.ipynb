## make `Collection_Date` values the same format

```python
# see dtype and existing values
sequence['Collection_Date'].head(20)
```

```python
# Step 1: Convert to datetime (year-month defaults to 1st, year-only becomes Jan 1st)
sequence['Collection_Date'] = pd.to_datetime(sequence['Collection_Date'], format='mixed', errors='coerce')
```

```python
# Step 2: Extract year and month for reference
sequence['Year'] = sequence['Collection_Date'].dt.year
sequence['Month'] = sequence['Collection_Date'].dt.month
```

```python
# Step 3: Identify rows that were originally just a year (month=1 and day=1)
sequence['Original_Day'] = pd.to_datetime(sequence['Collection_Date']).dt.day
```

```python
# Step 4: For each year, find the mode (most common) month from complete dates
# Then fill in year-only rows with that mode month

for year in sequence['Year'].unique():
    if pd.notna(year):  # Skip NaT years
        year_data = sequence[sequence['Year'] == year]
        
        # Find rows with complete dates (not defaulted to Jan 1st)
        complete_dates = year_data[(year_data['Month'] != 1) | (year_data['Original_Day'] != 1)]
        
        if len(complete_dates) > 0:
            # Get the mode month for that year
            mode_month = complete_dates['Month'].mode()
            
            if len(mode_month) > 0:  # If there is a mode
                mode_month = mode_month[0]
                
                # Update rows that were year-only (Month=1, Day=1)
                year_only_mask = (sequence['Year'] == year) & (sequence['Month'] == 1) & (sequence['Original_Day'] == 1)
                sequence.loc[year_only_mask, 'Collection_Date'] = pd.to_datetime(f'{int(year)}-{int(mode_month):02d}-01')
```

```python
# Step 5: Update Month column to reflect changes
sequence['Month'] = sequence['Collection_Date'].dt.month

# Step 6: Clean up temporary column
sequence = sequence.drop(columns=['Original_Day'])

# Check results
sequence[['Collection_Date', 'Year', 'Month']].head(20)
```

## Changing countries to regions for "Geo_Location"

```python
# see all countries we have
all_countries = sorted(sequence['Geo_Location'].dropna().unique())
for country in all_countries:
    print(country)
```

```python
import pycountry_convert as pc
import numpy as np

def get_continent(location):
    if pd.isna(location):
        return np.nan
    
    # Extract just the country name (before the colon)
    country = location.split(':')[0].strip()
    
    try:
        # Handle special cases
        if country == 'USA':
            return 'North America'
        elif country == 'UK' or country == 'United Kingdom':
            return 'Europe'
        elif country in ['Viet Nam', 'Vietnam']:
            return 'Asia'
        elif country == 'Antarctica':
            return 'Antarctica'
        elif country == 'Reunion':
            return 'Africa'
        elif country == "Cote d'Ivoire":
            return 'Africa'
        elif country == 'West Bank':
            return 'Asia'
        elif country == 'Kosovo':
            return 'Europe'
        elif country == 'Timor-Leste':
            return 'Asia'
        elif country == 'Borneo':
            return 'Asia'
        elif country == 'Korea':
            return 'Asia'
        elif country == 'Saint Barthelemy':
            return 'North America'
        elif country == 'Gaza Strip':
            return 'Asia'
        
        # Use the library for standard countries
        country_code = pc.country_name_to_country_alpha2(country)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent = pc.convert_continent_code_to_continent_name(continent_code)
        return continent
    except:
        return np.nan

# Apply the function
sequence['Continent'] = sequence['Geo_Location'].apply(get_continent)

# Check what didn't convert
unmapped = sequence[sequence['Continent'].isnull()]['Geo_Location'].value_counts()
print(f"Unmapped locations: {len(unmapped)}")
```
