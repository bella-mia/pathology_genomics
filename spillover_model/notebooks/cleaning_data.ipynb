## what each row title means

`Accession:` unique number associated with the data

`Organism_Name:` the organism that carries the virus

`Species:` the species of virus present in the organism

`Genotype:` the specific DNA sequence for the virus

`Isolate:` a specific strain of the virus that was collected from a specimen

`Length:` how long the virus's geome sequence is (how many nucleotides long)

`Nuc_Completeness:`whether or not nucleotide sequence is completed

`Geo_Location:` country or area where the virus strain was found

`Host:` animal where the virus strain was found

`Tissue_Speciman_Source:` where the virus tissue was found from the animal (feces, blood, saliva, tissue, etc)

`Submitters:` the people or institution who submitted the virus to the database

`Collection_Date:` the date the virus was collected from the host
