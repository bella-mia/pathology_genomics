```python
import random, math, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
import altair as alt
from vega_datasets import data as vega_data
import ipywidgets as widgets
from IPython.display import display, clear_output

warnings.filterwarnings('ignore')

BG=  '#e8f0fa'; CARD='#ffffff'; BORD='#c5d6ec'
P1=  '#2563eb'; P2=  '#60a5fa'; P3=  '#1e40af'
TXT= '#1e3a5f'; SEC= '#6b8aaa'; RED= '#dc2626'
YEL= '#d97706'; TEA= '#0891b2'; EXP= '#7c3aed'

plt.rcParams.update({
    'figure.facecolor':BG,   'axes.facecolor':CARD,  'axes.edgecolor':BORD,
    'text.color':TXT,        'axes.labelcolor':SEC,  'xtick.color':SEC,
    'ytick.color':SEC,       'grid.color':'#dde9f7', 'grid.linewidth':0.6,
    'legend.facecolor':CARD, 'legend.edgecolor':BORD,'font.family':'sans-serif',
})

AREA=50
DATA='spillover_model/data/processed/features.csv'

df = pd.read_csv(DATA, low_memory=False)
df['country'] = df['Geo_Location'].dropna().str.split(':').str[0].str.strip()

COUNTRY_ISO = {
    'Afghanistan':4,'Albania':8,'Algeria':12,'Angola':24,'Argentina':32,'Armenia':51,
    'Australia':36,'Austria':40,'Azerbaijan':31,'Bangladesh':50,'Belgium':56,'Benin':204,
    'Bolivia':68,'Brazil':76,'Bulgaria':100,'Burkina Faso':854,'Burundi':108,'Cambodia':116,
    'Cameroon':120,'Canada':124,'Central African Republic':140,'Chile':152,'China':156,
    'Colombia':170,'Congo':178,'Democratic Republic of the Congo':180,'Costa Rica':188,
    'Croatia':191,'Cuba':192,'Czech Republic':203,'Denmark':208,'Ecuador':218,'Egypt':818,
    'El Salvador':222,'Ethiopia':231,'Finland':246,'France':250,'Gabon':266,'Germany':276,
    'Ghana':288,'Greece':300,'Guatemala':320,'Guinea':324,'Honduras':340,'Hungary':348,
    'India':356,'Indonesia':360,'Iran':364,'Iraq':368,'Israel':376,'Italy':380,
    'Ivory Coast':384,'Jamaica':388,'Japan':392,'Jordan':400,'Kazakhstan':398,'Kenya':404,
    'Laos':418,'Lebanon':422,'Liberia':430,'Libya':434,'Madagascar':450,'Malawi':454,
    'Malaysia':458,'Mali':466,'Mexico':484,'Mongolia':496,'Morocco':504,'Mozambique':508,
    'Myanmar':104,'Nepal':524,'Netherlands':528,'New Zealand':554,'Nicaragua':558,
    'Nigeria':566,'Norway':578,'Pakistan':586,'Panama':591,'Paraguay':600,'Peru':604,
    'Philippines':608,'Poland':616,'Portugal':620,'Romania':642,'Russia':643,'Rwanda':646,
    'Saudi Arabia':682,'Senegal':686,'Sierra Leone':694,'Singapore':702,'South Africa':710,
    'South Korea':410,'Spain':724,'Sri Lanka':144,'Sudan':729,'Sweden':752,'Switzerland':756,
    'Taiwan':158,'Tanzania':834,'Thailand':764,'Tonga':776,'Trinidad and Tobago':780,
    'Turkey':792,'Uganda':800,'Ukraine':804,'United Kingdom':826,'USA':840,
    'United States':840,'Uruguay':858,'Uzbekistan':860,'Venezuela':862,'Vietnam':704,
    'Yemen':887,'Zambia':894,'Zimbabwe':716,
}

country_risk = (
    df[df['country'].notna() & ~df['host_group'].isin(['unknown','human'])]
    .groupby('country')
    .agg(mean_risk=('phylo_risk_score','mean'), sequence_count=('Accession','count'))
    .reset_index()
)
country_risk['id'] = country_risk['country'].map(COUNTRY_ISO)
country_risk = country_risk.dropna(subset=['id'])
country_risk['id'] = country_risk['id'].astype(int)

host_summary = (
    df[df['Host'].notna() & ~df['host_group'].isin(['unknown','human'])]
    .groupby('Host')
    .agg(count=('Accession','count'), phylo_risk=('phylo_risk_score','mean'), host_group=('host_group','first'))
    .reset_index()
)
host_summary['combined_score'] = host_summary['phylo_risk'] * np.log1p(host_summary['count'])
top_hosts = host_summary.sort_values('combined_score', ascending=False).head(15)

def run_seir(beta, gamma, sigma, N, I0, T):
    S,E,I,R = np.empty(T+1),np.empty(T+1),np.empty(T+1),np.empty(T+1)
    S[0],E[0],I[0],R[0] = N-I0,0.0,I0,0.0
    for t in range(T):
        ne=min(beta*S[t]*I[t]/N,S[t]); ni=min(sigma*E[t],E[t]); nr=min(gamma*I[t],I[t])
        S[t+1]=S[t]-ne; E[t+1]=E[t]+ne-ni; I[t+1]=I[t]+ni-nr; R[t+1]=R[t]+nr
    return np.column_stack([S,E,I,R])

rng=np.random.default_rng(42); risk_scores=np.linspace(0.05,0.95,30); seir_results=[]
for risk in risk_scores:
    betas=rng.gamma(2.0,(0.3+0.4*risk)/2,200); gammas=rng.gamma(2.0,0.10,200); sigmas=rng.gamma(2.0,0.20,200)
    runs=np.stack([run_seir(b,g,s,10000,1,160) for b,g,s in zip(betas,gammas,sigmas)])
    peak_I=runs[:,:,2].max(axis=1)
    seir_results.append({'risk_score':risk,'median_peak':np.median(peak_I),'lower':np.percentile(peak_I,25),'upper':np.percentile(peak_I,75)})
scatter_df=pd.DataFrame(seir_results)

host_points=pd.DataFrame([
    {'host':'Bat',     'risk':np.exp(-95/50),  'color':'#2563eb','dx': 8,'dy': 20},
    {'host':'Pig',     'risk':np.exp(-95/50),  'color':'#7c3aed','dx': 8,'dy':-28},
    {'host':'Rodent',  'risk':np.exp(-87/50),  'color':'#0891b2','dx':-55,'dy':12},
    {'host':'Pangolin','risk':np.exp(-80/50),  'color':'#3b82f6','dx': 8,'dy': 12},
    {'host':'Primate', 'risk':np.exp(-29/50),  'color':'#1e40af','dx': 8,'dy': 12},
    {'host':'Chimp',   'risk':np.exp(-6.9/50), 'color':'#1d4ed8','dx':-50,'dy':12},
])
host_points['median_peak']=host_points['risk'].apply(
    lambda r: scatter_df.iloc[(scatter_df['risk_score']-r).abs().argsort().iloc[0]]['median_peak'])

def make_global_map():
    world=alt.topo_feature(vega_data.world_110m.url,'countries')
    bg=alt.Chart(world).mark_geoshape(fill='#d0e4f7',stroke='#fff',strokeWidth=0.5)
    choro=(alt.Chart(world).mark_geoshape(stroke='#fff',strokeWidth=0.4)
        .transform_lookup(lookup='id',from_=alt.LookupData(country_risk,'id',['mean_risk','sequence_count','country']))
        .encode(
            color=alt.condition('datum.mean_risk != null',
                alt.Color('mean_risk:Q',scale=alt.Scale(scheme='blues',domain=[0.2,0.6]),
                    legend=alt.Legend(title='Phylo Risk',titleColor=TXT,labelColor=SEC,gradientLength=100,titleFontSize=10,labelFontSize=9)),
                alt.value('#d0e4f7')),
            tooltip=[alt.Tooltip('country:N',title='Country'),alt.Tooltip('mean_risk:Q',title='Risk Score',format='.3f'),alt.Tooltip('sequence_count:Q',title='Sequences',format=',')]
        ))
    return ((bg+choro).project('naturalEarth1')
        .properties(width=860,height=390,title='Global Spillover Risk by Country')
        .configure(background=BG)
        .configure_title(fontSize=13,fontWeight='bold',color=TXT,anchor='start',dy=-5)
        .configure_view(strokeWidth=0,fill=BG))

HOST_COLORS={'primate':'#1e40af','bat':'#2563eb','rodent':'#3b82f6','carnivore':'#1d4ed8',
             'pig':'#60a5fa','pangolin':'#2980c9','bird':'#93c5fd','cattle':'#4a9de0','other':'#bfdbfe'}

def make_top_hosts_fig():
    fig,ax=plt.subplots(figsize=(11,5.5),facecolor=BG); ax.set_facecolor(CARD)
    bar_cols=[HOST_COLORS.get(g,P1) for g in top_hosts['host_group']]
    ax.barh(top_hosts['Host'],top_hosts['combined_score'],color=bar_cols,edgecolor=BORD,linewidth=0.5,height=0.72)
    ax.invert_yaxis()
    ax.set_title('Top 15 Highest-Risk Host Species\n(phylogenetic risk x sequence volume)',color=TXT,fontsize=12,fontweight='bold',pad=10)
    ax.set_xlabel('Combined Risk Score',color=SEC,fontsize=10); ax.tick_params(labelsize=9)
    for sp in ax.spines.values(): sp.set_color(BORD)
    ax.grid(axis='x',alpha=0.5,lw=0.7)
    legend_els=[mpatches.Patch(facecolor=c,label=g.capitalize(),edgecolor=BORD) for g,c in HOST_COLORS.items() if g in top_hosts['host_group'].values]
    ax.legend(handles=legend_els,loc='lower right',fontsize=8,facecolor=CARD,edgecolor=BORD)
    plt.tight_layout(pad=1.2); return fig

def make_seir_fig():
    fig,ax=plt.subplots(figsize=(11,5.5),facecolor=BG); ax.set_facecolor(CARD)
    ax.fill_between(scatter_df['risk_score'],scatter_df['lower'],scatter_df['upper'],alpha=0.15,color=P1,label='25-75th percentile')
    ax.plot(scatter_df['risk_score'],scatter_df['median_peak'],color=P1,lw=2.5,label='Median peak infections',solid_capstyle='round')
    for _,row in host_points.iterrows():
        ax.scatter(row['risk'],row['median_peak'],color=row['color'],s=100,zorder=5,edgecolors='white',linewidths=0.8)
        ax.annotate(row['host'],xy=(row['risk'],row['median_peak']),xytext=(row['dx'],row['dy']),textcoords='offset points',
                    fontsize=9,color=row['color'],fontweight='bold',arrowprops=dict(arrowstyle='-',color=row['color'],lw=0.8))
    ax.set_title('Phylogenetic Risk Score vs Peak Infections\n(higher host-human relatedness = larger outbreaks)',color=TXT,fontsize=12,fontweight='bold',pad=10)
    ax.set_xlabel('Phylogenetic Risk Score  (0=distant, 1=close to humans)',color=SEC,fontsize=10)
    ax.set_ylabel('Median Peak Infected (out of 10,000)',color=SEC,fontsize=10)
    ax.tick_params(labelsize=9)
    for sp in ax.spines.values(): sp.set_color(BORD)
    ax.grid(axis='y',alpha=0.5,lw=0.7); ax.legend(fontsize=9,facecolor=CARD,edgecolor=BORD)
    plt.tight_layout(pad=1.2); return fig

# ── Harder simulation engine ──────────────────────────────────────────────────
class Agent:
    __slots__=['id','x','y','age','lifetime','zombie','exposed','exposed_timer','safe']
    def __init__(self,id,x,y,age=0,lifetime=10,zombie=False):
        self.id=id;self.x=x;self.y=y;self.age=age;self.lifetime=lifetime
        self.zombie=zombie;self.exposed=False;self.exposed_timer=0;self.safe=False

class SafeZone:
    def __init__(self,x,y,radius=5,strength=5,capacity=8):
        self.x=x;self.y=y;self.radius=radius;self.strength=strength;self.capacity=capacity

def _d(a,b): return math.hypot(a.x-b.x,a.y-b.y)

def init_pop(size=30,lifetime=11,zombies=2):
    pop=[Agent(i,random.uniform(0,AREA),random.uniform(0,AREA),age=random.randint(0,lifetime-1),lifetime=lifetime) for i in range(size)]
    for z in random.sample(pop,min(zombies,len(pop))): z.zombie=True
    return pop

def move_agents(pop,zones):
    zone_occ={id(z):sum(1 for a in pop if a.safe and not a.zombie) for z in zones}
    n_z=sum(1 for a in pop if a.zombie)
    zombie_speed=min(2.2+n_z*0.015, 3.5)
    for a in pop:
        if a.zombie:
            tgts=[h for h in pop if not h.zombie and not h.exposed]
            if not tgts: dx,dy=random.uniform(-1,1),random.uniform(-1,1)
            else: t=min(tgts,key=lambda h:_d(a,h)); dx,dy=t.x-a.x,t.y-a.y
            m=math.hypot(dx,dy)+1e-9
            a.x=max(0,min(AREA,a.x+dx/m*zombie_speed)); a.y=max(0,min(AREA,a.y+dy/m*zombie_speed))
        else:
            spd=0.8 if a.exposed else 1.2
            if zones:
                z=min(zones,key=lambda z:math.hypot(a.x-z.x,a.y-z.y))
                dx,dy=z.x-a.x,z.y-a.y; d=math.hypot(dx,dy)
                if d<=z.radius and zone_occ.get(id(z),0)<z.capacity:
                    a.safe=True; zone_occ[id(z)]=zone_occ.get(id(z),0)+1
                elif d>z.radius:
                    m=d+1e-9; a.x=max(0,min(AREA,a.x+dx/m*spd)); a.y=max(0,min(AREA,a.y+dy/m*spd))
            else:
                dx,dy=random.uniform(-1,1),random.uniform(-1,1)
                a.x=max(0,min(AREA,a.x+dx*spd)); a.y=max(0,min(AREA,a.y+dy*spd))

def infect_agents(pop,zones,radius=3.0,chance=0.65):
    for a in pop:
        if a.zombie:
            for o in pop:
                if not o.zombie and not o.exposed and not o.safe and _d(a,o)<=radius and random.random()<chance:
                    o.exposed=True; o.exposed_timer=random.randint(3,6)

def progress_exposed(pop):
    for a in pop:
        if a.exposed:
            a.exposed_timer-=1
            if a.exposed_timer<=0:
                a.exposed=False; a.zombie=True; a.safe=False

def attack_zones(pop,zones):
    breached,remove=0,[]
    for za in [a for a in pop if a.zombie]:
        for z in zones:
            if math.hypot(za.x-z.x,za.y-z.y)<=z.radius+1 and random.random()<0.18: z.strength-=1
            if z.strength<=0 and z not in remove: remove.append(z); breached+=1
    for z in zones:
        for a in pop:
            if a.exposed and a.safe and a.exposed_timer<=1 and math.hypot(a.x-z.x,a.y-z.y)<=z.radius+1:
                z.strength-=2
        if z.strength<=0 and z not in remove: remove.append(z); breached+=1
    for z in remove:
        zones.remove(z)
        for h in pop: h.safe=False
    return breached

def reproduce_pop(pop,cc=250):
    new_a,nid=[],max(a.id for a in pop)+1 if pop else 0
    humans=[a for a in pop if not a.zombie and not a.exposed]
    p=max(0.03,1-len(pop)/cc)
    for a in humans:
        nb=[o for o in humans if o!=a and _d(a,o)<=8]
        if nb and random.random()<p:
            nx=max(0,min(AREA,a.x+random.uniform(-1,1))); ny=max(0,min(AREA,a.y+random.uniform(-1,1)))
            new_a.append(Agent(nid,nx,ny,lifetime=a.lifetime)); nid+=1
    return new_a

def sim_step(pop,zones,kill=0.0,food=1.0,inf_r=3.0,inf_c=0.65):
    move_agents(pop,zones); infect_agents(pop,zones,radius=inf_r,chance=inf_c)
    progress_exposed(pop); breached=attack_zones(pop,zones)
    babies=reproduce_pop(pop); pop.extend(babies)
    dead=[]
    for a in pop:
        if a.zombie:
            if random.random()<kill: dead.append(a)
        elif not a.exposed:
            if a.age>=a.lifetime or random.random()<0.02*(1/max(food,0.1)): dead.append(a)
        a.age+=1
    for d in dead:
        if d in pop: pop.remove(d)
    h=sum(1 for a in pop if not a.zombie and not a.exposed)
    e=sum(1 for a in pop if a.exposed)
    z=sum(1 for a in pop if a.zombie)
    return h,e,z,breached

COOLDOWNS={'vaccinate':5,'quarantine':3,'supply':4}

class SimState:
    def __init__(self): self.reset()

    def reset(self,pop_size=30,phylo_risk=5,n_zones=2):
        self.step=0; self.game_over=False; self.outcome=''
        self.phylo_risk=phylo_risk
        self.inf_radius=1.8+phylo_risk*0.35; self.inf_chance=0.40+phylo_risk*0.055
        init_z=max(1,round(phylo_risk/2.5))
        self.population=init_pop(size=pop_size,zombies=init_z)
        zc={1:[SafeZone(25,25,radius=6,strength=6,capacity=10)],
            2:[SafeZone(13,13,radius=4,strength=5,capacity=8),SafeZone(37,37,radius=4,strength=5,capacity=8)],
            3:[SafeZone(13,13,radius=3,strength=4,capacity=6),SafeZone(37,13,radius=3,strength=4,capacity=6),SafeZone(25,37,radius=3,strength=4,capacity=6)]}
        self.zones=zc.get(n_zones,zc[2])
        self.hist={'t':[],'h':[],'e':[],'z':[],'r0':[]}
        self.cooldowns={k:0 for k in COOLDOWNS}
        self.vaccine_doses=max(2,6-phylo_risk//2)
        self.msg='Pathogen detected. Purple = incubating (hidden threat). Intervene before zones are breached.'
        self.msg_type='info'
        self._record()

    def _record(self):
        h=sum(1 for a in self.population if not a.zombie and not a.exposed)
        e=sum(1 for a in self.population if a.exposed)
        z=sum(1 for a in self.population if a.zombie)
        self.hist['t'].append(self.step); self.hist['h'].append(h); self.hist['e'].append(e); self.hist['z'].append(z)
        r0=round(self.hist['z'][-1]/self.hist['z'][-3],2) if len(self.hist['z'])>=3 and self.hist['z'][-3]>0 else 0.0
        self.hist['r0'].append(r0)

    def advance(self,action='none'):
        if self.game_over: return
        if action in self.cooldowns and self.cooldowns[action]>0:
            self.msg=f'Cooldown: {action.capitalize()} available in {self.cooldowns[action]} step(s).'; self.msg_type='neutral'
            self._tick(); return
        kill,food,old_ic=0.0,random.uniform(0.6,1.4),self.inf_chance
        if action=='vaccinate':
            if self.vaccine_doses<=0:
                self.msg='No vaccine doses left! Use Supply to restock.'; self.msg_type='neutral'; self._tick(); return
            kill=0.28; food=1.0; self.vaccine_doses-=1
            self.msg=f'Vaccination deployed (28% cull). {self.vaccine_doses} dose(s) remaining.'; self.msg_type='success'
            self.cooldowns['vaccinate']=COOLDOWNS['vaccinate']
        elif action=='quarantine':
            self.inf_chance=max(0.05,self.inf_chance*0.25); food=0.75
            self.msg='Hard quarantine: transmission cut 75%, food supply strained.'; self.msg_type='warning'
            self.cooldowns['quarantine']=COOLDOWNS['quarantine']
        elif action=='supply':
            food=3.0
            self.vaccine_doses=min(self.vaccine_doses+1, max(2,6-self.phylo_risk//2))
            self.msg='Medical supplies deployed: mortality reduced, +1 vaccine dose restocked.'; self.msg_type='info'
            self.cooldowns['supply']=COOLDOWNS['supply']
        else:
            self.msg='No intervention — pathogen spreads unchecked.'; self.msg_type='neutral'
        h,e,z,breached=sim_step(self.population,self.zones,kill=kill,food=food,inf_r=self.inf_radius,inf_c=self.inf_chance)
        if action=='quarantine': self.inf_chance=old_ic
        self.step+=1; self._record(); self._tick()
        if self.step>0 and self.step%random.randint(8,13)==0 and random.random()<0.45:
            self.inf_radius=min(self.inf_radius*1.3,AREA*0.4); self.inf_chance=min(self.inf_chance*1.2,0.95)
            self.msg+='  MUTATION: pathogen evolved, higher transmissibility!'; self.msg_type='danger'
        if breached: self.msg+=f'  {breached} containment zone(s) breached!'
        if h+e==0 or (h==0 and z>0):
            self.game_over=True; self.outcome='failure'; self.msg='Host population extinct. Pathogen wins.'
        elif z==0 and e==0:
            self.game_over=True; self.outcome='success'; self.msg='Pathogen fully contained! Humanity survives.'
        elif self.step>=80:
            self.game_over=True; self.outcome='partial'; tot=h+e+z; self.msg=f'Simulation ended. Survival: {h/tot*100:.0f}%'

    def _tick(self):
        for k in self.cooldowns:
            if self.cooldowns[k]>0: self.cooldowns[k]-=1

sim=SimState()
```
